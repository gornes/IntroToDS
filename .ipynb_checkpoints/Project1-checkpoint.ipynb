## Loading the improved data set.

We load the data set from the given csv file into a Pandas DataFrame instance. The data is already *clean* and ready for data analysis, however some extra "wrangling" is performed: 

* After loading the data, we create a `dateTime` column that transform the string `datetime` column into a `pd.datetime` variable. 

* Also, an `hour_week` column is created: this column represents the *hours of the week*, an integer with a value between 0 and 167, where 0 is Monday at 00:00 hours and 167 is Sunday at 23:00 hours.

```python
improved_df = pd.read_csv(
    '/home/itoledo/Dropbox/OnlineClass/Udacity/IntroToDataScience/IntroToDS/turnstile_weather_v2.csv')
improved_df['dateTime'] = improved_df.apply(
    lambda r: pd.datetime.strptime(r['datetime'], '%Y-%m-%d %H:%M:%S'), axis=1)
improved_df = improved_df.set_index('dateTime', drop=False)
improved_df['hour_week'] = improved_df.day_week * 24. + improved_df.hour
improved_df['day'] = improved_df.index.day
improved_df['week'] = improved_df.index.week
print "The number of entries in the data set is: %d" % len(improved_df)
print "The number of turnstiles is: %d" % len(improved_df.UNIT.unique())
```

We check now for the uniformity of monitoring points of the turnstiles. Are all turnstiles reporting every four hours though the month of May? To answer this question let's start with the distribution of the number of data reported for each turnstile.

We found out that several turnstiles have missing data during the studied time range, with only 52 stations having all data. Can we keep using the improved data set as it is? Three questions arise:

* Are the missing entries randomly distributed in terms of date, time and station for these turnstiles?

* If not, could this issue contribute with a selection bias?

* Should we dismiss these data for the analysis of fill in the missing data?


After studying the 3 clear outliers, units R260, R295 and R459, we decide to got rid of these because it was clear that they had too many values missing or equal to 0. It is imposible to know from the available data if the missing data is related to failed reports, closed stations or other problems; further more, including these stations will introduce deformations in our statistics, since the ridership, when available, is in many cases the number of entries since the last good point, producing misleading "ridership picks" on certains hours, i.e: if there was no report in the last 48 hours, the next value reported is the accumulation of entries since the last valid entry.

```python
# Remove outlier turnstiles
improved_df = improved_df.query('UNIT not in ["R260", "R295", "R459"]')
```

### Checking for consitency between the "rain" value and variables precipi, meanprecipi and conds.

```python
improved_df.head(2)
```

From the exploration of the data, and disccussion at the forum, we have realized that the `rain` column might have inconsistencies. There are four relevant columns for precipitation conditions: `rain`, `precipi`, `meanprecipi` and `conds`:

* `rain`: Indicator (0 or 1) if rain occurred withing the **calendar day** at the location
* `precipi`: Precipitation in inches at the **time** and location
* `meanprecipi`: **Daily** average of `precipi` for the location
* `conds`: Categorical variable of the weather conditions (Clear, Cloudy, etc.) for the **time** and location

However it was noticed that inconsistencies are present between `rain` and `meanprecipi`. This is how we get 93 (by weather station location) entries with days marked as **rainy** but the mean precipitation for the day is 0:

```python
t = improved_df.groupby(['weather_lat', 'datetime', 'conds'],
                   as_index=False).max()[
    ['weather_lat', 'datetime', 'conds', 'rain', 'precipi', 'meanprecipi']]
t[(t.rain == 1) & (t.meanprecipi == 0.)].__len__()
```

What happens is that a whole day at a weather station location is marked as **rainy** if at least once during the day the conditions where Rain, Hard Rain, Light Rain or Drizzle, and some `precipi` values were missing and replaced by 0.

This means that we can't use the time variable `precipi`, but instead we can use `conds` to check hourly conditions and daily conditions.

Still, this will limit our analysis either if we use the daily condition or the hourly condition. Let's show the problems our data will have by means of an example with turnstile `R084`:

```python
r84 = improved_df.query('UNIT == "R084"').groupby(['weather_lat', 'datetime', 'conds'],
                   as_index=False).max()[
    ['weather_lat', 'datetime', 'conds', 'rain', 'precipi', 'meanprecipi','ENTRIESn_hourly']]
r84['mhour'] = r84.index.values
r84_1 = r84[(r84.rain == 1)]
r84_0 = r84[(r84.rain == 0)]
```

```python
plt.figure(figsize=[18,7])
pts = plt.scatter(r84_1.mhour, r84_1.ENTRIESn_hourly, c=r84_1['precipi'].values, cmap='Blues', s=60, label='Rainy data')
plt.scatter(r84_0.mhour, r84_0.ENTRIESn_hourly, c=r84_0['precipi'].values, cmap='Blues', marker='^', label='Non-rainy data', s=3)
plt.plot(r84.mhour, r84.ENTRIESn_hourly, 'k-', lw=0.2)
plt.xlim(0,187)
plt.ylim(0,35000)
plt.colorbar(pts)
plt.legend(fontsize=13)
plt.xlabel('Time index', fontsize=13)
plt.ylabel('Hourly ridership', fontsize=13)
plt.title('Ridership and precipitation for Turnstile R084', fontsize=16)
plt.savefig(figpath + 'r084.png', dpi=180)
```
